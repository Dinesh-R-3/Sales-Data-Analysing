## Top 10 Product Names by store, in terms of Value for Jan, Feb, and March

## STORE A_

```python
store_a_jan = dataset.loc[(dataset['Month_name'] == 'January')&(dataset['Store ID']=='Store A')]
```

```python
print(store_a_jan)
```

```python
store_a_feb = dataset.loc[(dataset['Month_name'] == 'February')&(dataset['Store ID']=='Store A')]
store_a_mar = dataset.loc[(dataset['Month_name'] == 'March')&(dataset['Store ID']=='Store A')]
```

```python
top_10_jan=store_a_jan.groupby(["Product Name"])["Value"].sum().sort_values(ascending=False).head(10)
```

```python
print(top_10_jan)
```

```python
top_10_feb=store_a_feb.groupby(["Product Name"])["Value"].sum().sort_values(ascending=False).head(10)
top_10_mar=store_a_mar.groupby(["Product Name"])["Value"].sum().sort_values(ascending=False).head(10)
```

```python
print(top_10_feb)
print(top_10_mar)
```

```python
sns.set_style("darkgrid")
```

```python
def sales_plots(dataset,col_name):
    plt.figure(figsize=(14,4))
    sns.barplot(x="Product Name",y="Value",data=dataset,palette="rocket").set_title([col_name],
                                                                               color="DarkRed",fontsize=20)
    plt.xticks(rotation=90)
    plt.xlabel("Product_Names",fontsize=15,color="Blue")
    plt.ylabel("Sales_value",fontsize=15,color="Blue")
    plt.show()
```

```python
print(top_10_jan)
```

```python
Store_A_January=pd.DataFrame(top_10_jan).reset_index()
Store_A_January
```

```python
sales_plots(Store_A_January,"Store A / January")
```

```python
dataset.head()
```

#### STORE A - February sales

```python
Store_A_February=pd.DataFrame(top_10_feb).reset_index()
Store_A_February
```

```python
sales_plots(Store_A_February,"Store A/February")
```

#### STORE A - March sales

```python
Store_A_March=pd.DataFrame(top_10_mar).reset_index()
Store_A_March
```

```python
sales_plots(Store_A_March,"Store_A/March")
```

- Combine Sales for STORE A for jan,feb & mar

```python
Store_A_January=Store_A_January.rename({"Value":"value/_January"},axis=1)
```

```python
Store_A_Febraury=Store_A_February.rename({"Value":"value/_February"},axis=1)
```

```python
Store_A_March=Store_A_March.rename({"Value":"value/_March"},axis=1)
```

```python
sales_store_A=pd.concat([Store_A_January,Store_A_Febraury,Store_A_March],axis=1)
```

```python
sales_store_A
```
